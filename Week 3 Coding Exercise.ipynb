 - Create a variable to keep track of the running npv total as you <br>calculate the value of the cashflows (i.e. call the variable <i>total</i>)
 - Request the number of cashflow amounts
 - Request the discount rate
 - Create a "For loop" that loops the same number of times as there are cashflow items.<br>
 (i.e. <i><b>for i in range (1,numberOfCashflowAmounts):</b></i>
 - Within the loop keep the running total/value 
 <br>(i.e. <i><b>total = total + (cashflowAmount / (1 + rate) ** i</b> </i> )<br>
 (where i is the iteratin variable)
 - After the loop ends print the <i>total</i> variable within a nicely formated message


```python
numberofcashflowamt = int(input("Enter the number of cashflow values: "))

discountrate = float(input("Enter the period discount rate: "))

sum = 0
for i in range(1,numberofcashflowamt+1):
    cashflowperyear = input("Please enter cashflow #"+ str(i) + ":")
    pv = float((cashflowperyear)) / float((1+discountrate)**i)
    sum += pv
    
print("The NPV for this cash flow is " + str(sum))


```
